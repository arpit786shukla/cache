```python
import pandas as pd
import numpy as np
from collections import defaultdict 
import sys
import os
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
ADDR = '/nfs_home/nbhardwaj/data/rds_data/400/'
```

```python
sets = [x for x in range(64)]
n_wrong = []
n_correct = []
n_total = []
m_len = []
for cset in sets:
    df = pd.read_csv(ADDR+str(cset)+'.csv', index_col = [0])
    n_total.append(len(df))
    m = {}
    pos, neg = 0, 0
    for x, y in df[['Instruction', 'label']].values:
    #     print(x, y)
        if(x in m.keys() and m[x]==y):
            pos += 1
        else:
            neg += 1
            m[x] = y
    m_len.append(len(m))
    n_correct.append(pos)
    n_wrong.append(neg)
```

```python
n_wrong = np.asarray(n_wrong)
n_correct = np.asarray(n_correct)
n_total = np.asarray(n_total)
acc = np.divide(n_correct, n_total)
plt.plot([x for x in range(len(acc))], 100*acc)
plt.xlabel('set no')
plt.ylabel('accuracy')
plt.title('Baseline model') 








          
plt.show()
```

```python
pos = np.sum(n_correct)
tot = np.sum(n_total)
print(pos*100/tot)
```
